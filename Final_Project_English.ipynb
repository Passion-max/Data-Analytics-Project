## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.
data.isnull().sum().sort_values()
```

```python
def fill_missing_values(col):
    if col.dtype == 'object':
        return col.fillna(col.mode()[0])
    else:
        return col.fillna(col.median())

data = data.apply(fill_missing_values)
```

```python
data.isnull().sum().sort_values()
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.
number_of_duplicates = data.duplicated().sum()
number_of_records_Old = data.shape

print(f"Number of Duplicates in the Dataset is {number_of_duplicates}")
print(f"Number of records in the dataset before removing duplicate is {number_of_records_Old}")

data.drop_duplicates(inplace=True)
```

```python
## Display the total number of records in the dataframe after removing the duplicates.

number_of_records_new = data.shape

print(f"The number of records after removing duplicates is {number_of_records_new}")
```

## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`
- Convert all column names to lowercase and replace the spaces in the column names with an underscore "_".
- Remove the dollar sign and comma from the columns `price` and `service_fee`. If necessary, convert these two columns to the appropriate data type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python


## Rename the column.
data = data.rename(columns={"availability 365": "days_booked"})
```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"
data.columns = [col.lower().replace(" ", "_") for col in data.columns]
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.
def transform_value(value):
    return float(value.replace("$", "").replace(" ", "").replace(",", ""))

# Apply the function to the 'price' and 'service_fee' columns
data['price'] = data['price'].apply(transform_value)
data['service_fee'] = data['service_fee'].apply(transform_value)
data['construction_year'] = data['construction_year'].astype(int)
```

```python
negative_nights = data[data['minimum_nights'] < 0]
print("Number of entries with negative minimum nights:", len(negative_nights))
```

```python
#I observe that there are negative values (13) in the dataset for minimum_nights columns so I created a new column call minimum_night_abs
data['minimum_nights_abs'] = data['minimum_nights'].abs()
```

```python
# data.to_csv('airbnb_transform.csv')
plt.figure(figsize=(10, 50))

short_term_neighborhoods = data[['neighbourhood', 'minimum_nights_abs']][data['minimum_nights_abs'] <= 10]
short_term_neighborhoods = short_term_neighborhoods['minimum_nights_abs'].groupby(short_term_neighborhoods['neighbourhood'])\
                            .mean().reset_index().sort_values(by=['minimum_nights_abs'], ascending=True)
short_term_neighborhoods['minimum_nights_abs'] = short_term_neighborhoods['minimum_nights_abs'].round().astype(int)

sns.barplot(y='neighbourhood', x='minimum_nights_abs', data=short_term_neighborhoods)


plt.title("Neighbourhood With less than 10 Minimum Nights")
plt.show()
```

```python
## List the prices with respect to room type using a bar graph and also state your inferences.
price_per_room_type = data['price'].groupby(data['room_type']).mean().sort_values(ascending=False).reset_index()


plt.figure(figsize=(10, 6))
sns.barplot(y='price', x='room_type', data=price_per_room_type)


plt.title("Price by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price")


plt.ylim(600, 700)
plt.show()
```

## Inference from the above graph

### Range of Prices:

All room types have average prices within a relatively narrow range between approximately `620` to `666`. This suggests that there's not a significant price variation among different room types in the dataset.

### Most Expensive and Least Expensive Rooms: 

`Hotel room` is the **most expensive room type on average**, while `Private room` is the **least expensive**. This might reflect a general trend where hotel rooms typically offer more amenities or privacy compared to other types of accommodations, justifying a higher price. On the other hand, private rooms, likely being a part of a larger residence, are less costly.

### Shared Rooms vs. Entire Homes/Apartments:

Interestingly, `Shared room` is the second most expensive category, even more than `Entire home/apt`. This could indicate that these shared rooms are in more desirable locations or offer unique amenities. Alternatively, it might reflect a small sample size or outliers in the data for shared rooms.

### Market Homogeneity: 

The close range of prices also suggests a homogeneous market for these accommodation types, where customers can expect similar pricing across different room types. This might imply a competitive market where no single type of room drastically undercuts or overshadows others in terms of pricing.

```python
## Create a pie chart that shows distribution of booked days for each neighborhood group .Which neighborhood has the highest booking percentage
booked_days_distribution = data['days_booked'].groupby(data['neighbourhood_group']).sum()

plt.figure(figsize=(10, 10))  
wedges, texts, autotexts = plt.pie(booked_days_distribution, autopct='%1.1f%%', startangle=140)

plt.legend(wedges, labels=booked_days_distribution.index, title="Neighbourhood Groups", loc="upper right")

plt.title("Distribution of Booked Days by Neighbourhood Group")

plt.show()
```

## Task 5b: Data Visualization (Any Tool)

* Does service price and room price have an impact on each other. Illustrate this relationship with a scatter plot and state your inferences
* Using a line graph show in which year the maximum construction of rooms took place.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
#Does service price and room price have an impact on each other. Illustrate this relationship with a scatter plot and state your inferences

plt.figure(figsize=(10, 5))

sns.scatterplot(x='price', y='service_fee', data=data, size=20, hue='room_type')

plt.title('Relationship Between Service Fee and Room Price')

plt.show()
```

#### Inferences

**Strong Linear Relationship:**
The plot shows a very strong linear relationship between `price` and `service_fee`. 
This implies that the service fee increases proportionally with the price, 
which might suggest a fixed percentage fee or a strong linear pricing strategy.

**Flat Line Along the X-axis:** 
There is a concentration of points along the x-axis at what seems to be a low `service_fee` across various prices.
This could indicate a minimum service fee or a base fee that is consistent regardless of the price.

```python
## Using a line graph show in which year the maximum construction of rooms took place.
#convert 'construction_year' from float to string for better display

yearly_construction  = data['construction_year'].groupby(data['construction_year']).count()

plt.figure(figsize=(15, 6))
plt.plot(yearly_construction.index.astype(str), yearly_construction.values, marker='o')

plt.title("Yearly Room Construction")
plt.xlabel("Year")
plt.ylabel("Number of Rooms Constructed")

plt.show()
```

#### Year with maximum construction - 2012 with 5150 Rooms
